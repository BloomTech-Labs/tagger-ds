```python
ORG_EMAIL = '@gmail.com'
FROM_EMAIL = '' + ORG_EMAIL
FROM_PWD = ''
SMTP_SERVER = 'imap.gmail.com'
SMTP_PORT = 993
```

```python
# Utility to read email from Gmail using Python
def read_email_from_gmail():
    msgs = {}
    try:
        mail = imaplib.IMAP4_SSL(SMTP_SERVER)
        mail.login(FROM_EMAIL, FROM_PWD)
        mail.select('inbox')
        type, data = mail.search(None, 'ALL')
        
        mail_ids = data[0]
        id_list = mail_ids.split()
        first_email_id = int(id_list[0])
        latest_email_id = int(id_list[-1])
        
        for i in range(latest_email_id, latest_email_id-1000, -1):
            typ, data = mail.fetch(str(i).encode(), '(UID BODY[TEXT])')
            typ2, data2 = mail.fetch(str(i).encode(), '(RFC822)')
            print(i)
            
            for response_part in data:
                if isinstance(response_part, tuple):
                    try:
                        content = response_part[1].decode()
                    except:
                        content = None
                        
            for response_part in data2:
                if isinstance(response_part, tuple):
                    try:
                        msg = email.message_from_string(response_part[1].decode())
                    except:
                        continue
                        #print('Message Content: ', content)
                msgs[i] = {'from': msg['from'], 'subject': msg['subject'],
                          'body': content}
    except Exception as inst:
        raise
    return msgs
```
