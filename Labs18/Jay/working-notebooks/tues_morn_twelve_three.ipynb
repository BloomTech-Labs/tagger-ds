```python
import re
# use regex to extract domain name of email address
domains = []
pattern = '\@(.*?)\.'
for val in df['sender_email']:
    match = re.findall(pattern, val)
    domains.append(match)
df['domain_name'] = [' '.join(s) for s in domains]
```

```python
df['domain_name'].value_counts()
```

```python
links = []
pattern = 'https?'
for val in df['Message']:
    match = re.findall(pattern, val)
    links.append(match)
    
df['link_count'] = [len(l) for l in links]
df['link_count'].value_counts()
```
