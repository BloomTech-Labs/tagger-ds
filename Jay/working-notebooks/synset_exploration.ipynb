```python
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
nltk.download('all')
```

```python
wn = wordnet
```

```python
query = 'travel home itinerary'
tokens = word_tokenize(query)
```

```python
for token in tokens:
    print(wn.synsets(token))
    print('-----------')
```

```python
sample = wn.synsets(tokens[0])

dir(sample)
```
